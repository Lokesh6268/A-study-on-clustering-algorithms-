```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("sonar.csv")
```

```python
# Step 1: Drop the label column
X = df.drop('Class', axis=1)
```

```python
# Step 2: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 3: Apply KMeans with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=25)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Step 4: Count samples in each cluster
print("Cluster counts:")
print(df['Cluster'].value_counts())
```

```python
# Step 5: Visualize the clusters
plt.figure(figsize=(6, 4))
df['Cluster'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title("Cluster Distribution (K=2)")
plt.xlabel("Cluster")
plt.ylabel("Number of Samples")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
