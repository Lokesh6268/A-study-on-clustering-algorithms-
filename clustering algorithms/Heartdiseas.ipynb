```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the dataset
df = pd.read_csv("heart-disease-data (2).csv")
```

```python
# Step 2: Drop the target column (we won't use it in clustering)
X = df.drop("target", axis=1)
```

```python
# Step 3: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 4: Apply KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=12)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Step 5: Show the number of samples per cluster
print("Cluster Counts:")
print(df['Cluster'].value_counts())
```

```python
# Step 6: Visualize the cluster distribution
plt.figure(figsize=(6, 4))
df['Cluster'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title("Cluster Distribution (k=2)")
plt.xlabel("Cluster")
plt.ylabel("Number of Samples")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
